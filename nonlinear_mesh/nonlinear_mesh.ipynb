```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton_krylov
from time import clock as clock
```

Testing nonlinear mesh with diffusion equation only and no flux boundary conditions (no source terms). We assume a constant diffusion constant.

Flux:
$$ f = -D \partial_x c(x,t) $$

Conservation law:
$$ \partial_t c(x,t) = - \partial_x f = - \partial_x \left( -D \partial_x c(x,t) \right)$$
$$ \partial_t c(x,t) = -\frac{1}{\Delta V_i} \left( f_{i+1} - f_i \right) $$

Spatial axis $ x \in [0,1] $ - one dimensional. Cell centers vector x

$$ f_{i+1} - f_i = -D_{i+1} \frac{2 \left(\alpha c_{i+1} - \beta c_i \right)}{\Delta V_{i+1} + \Delta V_i} $$

```python
# grid generation
I = 100

xi = np.arange(0.0, 1.0 + 1.0 / (I+1), 1.0 / (I) )

x = np.sin( ( np.pi * xi  ) / 2 ) ** 2
y = np.concatenate(([0.0], np.sort(np.random.rand(I-1)), [1.0]))

x = xi


# extract mitpoint
centers = np.zeros(I, dtype = np.float64)
for i in range(I):
    
    centers[i] = x[i] + (x[i+1] - x[i]) / 2

# time axis explicit integration
N = 1000
Dt = 0.001
time = np.zeros(N, dtype = np.float64 )
for j in range(0, N):
    
    time[j] = j * Dt

print(centers)

# nonlinear


print(y)

fig = plt.figure(dpi = 120)

ax = fig.add_subplot(1,1,1)

ax.plot(x, np.zeros(y.shape), ls = "None", marker = ".")
```

```python
# residual vector
def residual(sol_old, sol_new, Dx, D, I, Dt):
        
    res = np.zeros(I, dtype = np.float64)
    
    # anode boundary condition
    res[0] = (sol_new[0] - sol_old[0] 
            + Dt * (-D[1] * ( sol_new[1] - sol_new[0]) / ((Dx[0] + Dx[1]) / 2) ) / Dx[0])
    
    for i in range( 1, I-1 ):

        res[i] = ( sol_new[i] -sol_old[i] 
        + Dt * ( -D[i+1] * ( sol_new[i+1] -  sol_new[i]) / ((Dx[i+1] + Dx[i]) / 2.0) 
                       + D[i]  * ( sol_new[i] -  sol_new[i-1]) / ((Dx[i] + Dx[i-1]) / 2.0) ) / Dx[i] )
    
    # catode boundary condition
    res[I-1] = (sol_new[I-1]- sol_old[I-1] 
                + Dt * (D[I] * ( sol_new[I-1] - sol_new[I-2]) / ((Dx[I-1] + Dx[I-2]) / 2.0) ) / Dx[I-1])
    
    return res

# getting cell volumes
Dx = x[1:] - x[:I]

# init solution vector
sol = np.zeros( [I,N], dtype = np.float64 )

sol[:,0] = sol_initial

D = np.ones( I+1, dtype = np.float64 )

# time loop
for j in range(1, N):
    
    
    sol[:,j] = newton_krylov( lambda y: residual(sol[:,j-1], y, Dx, D, I, Dt), sol[:,j-1])
 
Dx = x[1:] - x[:I]

for j in range(N):
    solint[j] = np.dot(sol[:,j], Dx)
```

```python
# linear function as driving field defined also on cell centers
f = 2 * centers

# derivatives of cell interfaces

def flux_limiter( r ):
    """ van Leer flux limiter """
    
    if r <= 0:
        phi = 0.0
    else:
        #phi = 2.0 * r / ( 1.0 + r )
        phi = np.max([0.0, np.min([1.0, r])])
    return phi

def residual(sol_old, sol_new, f, Dx, D, I, Dt):
        
    res = np.zeros(I, dtype = np.float64)
    
    #
    # Godunov scheme
    #
    
    velup = - D[1] * (f[1] - f[0]) / ( (Dx[1] + Dx[0]) / 2.0) # velocity on right cell edge
    
    if velup >= 0:

        upflux = velup * (sol_new[0]  )

    else:

        upflux = sol_new[1] * velup
    
    # anode boundary condition
    res[0] = (sol_new[0] - sol_old[0] + Dt * (upflux) / Dx[0])

    # inner points
    for i in range(1,I-1):
        
        #
        # Godunov scheme - upwinding with (limited) piecewise constant reconstruction
        #
        
        # calculate velocity - driving electric field
        velup = - D[i+1] * (f[i+1] - f[i]) / ( (Dx[i+1] + Dx[i]) / 2.0) # velocity on right cell edge
        veldown = - D[i] * (f[i] - f[i-1]) / ( (Dx[i] + Dx[i-1]) / 2.0 ) # velocity on left cell edge
        
        # code flux limiter directly into the conditional statements
        
        # choosing limiter following wikipedia
        #r = ( sol_new[i+1] - sol_new[i] ) / ( sol_new[i] - sol_new[i-1] )
        r = ( sol_new[i] - sol_new[i-1] ) / (Dx[i] + Dx[i-1]) * (Dx[i+1] + Dx[i]) / ( sol_new[i+1] - sol_new[i] )
        
        if r <= 0.0:
            
            limiter = 0.0
            
        else:
            # van Leer
            #limiter = (r + np.abs(r)) / ( 1.0 + np.abs(r) )
            # minmod
            # limiter = np.max([0.0, np.min([1.0, r])])
            # superbee
            #limiter = np.max([0.0, np.min([2*r,1]), np.min([r, 2])])
            # MC
            limiter = np.max([0, np.min([2*r,0.5*(1+r),2])])
        
        # flux at rigth cell edge
        # for reconstruction: slope is 2 * (c_i+1 - c_i) / (Dx_i+1 + Dx_i)
        #
        if velup > 0:
            
            upflux = velup * ( sol_new[i] + Dx[i] * limiter * ( sol_new[i+1] - sol_new[i] ) / (Dx[i+1] + Dx[i]) )

        else:

            upflux = velup * ( sol_new[i+1] - Dx[i+1] * limiter * ( sol_new[i+1] - sol_new[i] ) / (Dx[i+1] + Dx[i]) )
        
        # flux at left cell edge
        if veldown > 0:

            downflux = veldown * ( sol_new[i-1] + Dx[i-1] * limiter * (sol_new[i] - sol_new[i-1]) / (Dx[i] + Dx[i-1]) )

        else:

            downflux = veldown * ( sol_new[i] - Dx[i] * limiter * (sol_new[i] - sol_new[i-1]) / (Dx[i] + Dx[i-1]) )
            
        res[i] = sol_new[i] -sol_old[i] + Dt * ( upflux - downflux ) / Dx[i]

    # velocity on f_left
    veldown = - D[I-1] * (f[I-1] - f[I-2]) / ( (Dx[I-1] + Dx[I-2]) / 2.0 )

    if veldown > 0:

        downflux = sol_new[I-2] * veldown

    else:

        downflux = sol_new[I-1] * veldown
        
    downflux = veldown * (sol_new[I-2] + sol_new[I-1]) / 2.0 - np.abs(veldown) * (sol_new[I-1] - sol_new[I-2]) / 2.0
    
    # catode boundary condition
    res[I-1] = (sol_new[I-1]- sol_old[I-1] + Dt * (- downflux ) / Dx[I-1])
    
    return res

# time axis explicit integration
N = 1000
Dt = 0.00008
time = np.zeros(N, dtype = np.float64 )
for j in range(0, N):
    
    time[j] = j * Dt

# getting cell volumes
Dx = np.zeros(I, dtype = np.float64)
Dx = x[1:] - x[:I]

# initial condition
sigma = 0.005
centers = np.zeros(I, dtype = np.float64)
centers = x[:I] + (x[1:] - x[:I])
sol_initial = np.exp( -(centers - 0.5) ** 2 / (2 * sigma)  ) / (sigma * np.sqrt(2*np.pi))

# init solution vector
sol = np.zeros( [I,N], dtype = np.float64 )

sol[:,0] = sol_initial

D = np.ones( I+1, dtype = np.float64 )

t1 = clock()
# time loop
for j in range(1, N):
    
    if np.mod(j,1) == 0:
        print("Time step :", j)
        
    sol[:,j] = newton_krylov( lambda y: residual(sol[:,j-1], y, f, Dx, D, I, Dt), sol[:,j-1])
     
t2 = clock()
    
print("Simulation Time: ", t2-t1)
Dx = x[1:] - x[:I]


solint = np.zeros(N, dtype = np.float64)
for j in range(N):
    solint[j] = np.dot(sol[:,j], Dx)
```
