```python
def solver( I, N, epsilon, DC, DA, sol_initial, chi2, M, reaction_bc ):
    
    # reactin boundary conditions
    kA = reaction_bc["kA"]
    kC = reaction_bc["kC"]
    foxA = reaction_bc["foxA"]
    foxC = reaction_bc["foxC"]
    
    # axis
    Dx, centers = calcAxis( I )
    
    current = np.zeros([2, N], dtype = np.float64)
    epsilon_vec = np.ones(I + 1, dtype = np.float64) * epsilon
    DC_vec = np.ones(I + 1, dtype = np.float64) * DC
    DA_vec = np.ones(I + 1, dtype = np.float64) * DA

    t1 = clock()

    sol = np.zeros([3 * I, N], dtype = np.float64)    

    sol[:,0] = sol_initial

    # calculate jacobian and invert it for the first points
    Jac = constJac(
                  I,
                  Dt,
                  Dx,
                  DC_vec,
                  DA_vec,
                  epsilon_vec,
                  chi2,
                  M
                  )

    Jacinv = np.linalg.inv(Jac)

    # delete Jacobian - only inverse J is needed
    #del Jac

    for j in range(1,N):

        if j <= 20:
            print("Time Step: ", j)
        elif j > 20 and np.mod(j,100) == 0:
            print("Time Step: ", j)
        
        if j < 5:
            
            t_step_method = 1
            
            sol2 = sol[:,j-1]
            
        else:
            
            t_step_method = 1
            
            sol2 = sol[:,j-2]
        
        
        sol[:,j] = newton_krylov( lambda y: residual(
                                    I,
                                    Dx,
                                    y,
                                    sol[:,j-1], # sol1 (j-1)
                                    sol2, # sol2 (j-2) for BDF2 only
                                    chi1,
                                    chi2,
                                    DC_vec,
                                    DA_vec,
                                    Dt,
                                    kA,
                                    kC,
                                    foxA,
                                    foxC,
                                    phiC[j],
                                    epsilon_vec,
                                    model,
                                    t_step_method,
                                    M
                                    ),
                                    sol[:,j-1],
                                    inner_M = Jacinv,
                                    method = "lgmres",
                                    verbose = 0,
                                    maxiter = 100)
        
        # start current calculation if j >= 2, when the system reacts properly
        if j >= 2:
            
            # anodic current
            current[0,j] = calcAnodicCurrent( sol[2*I,j], sol[2*I,j-1], sol[2*I,j-2], Dt, Dx[0], chi2 )
            
            # catodic current
            current[1,j] = calcCatodicCurrent( sol[3*I-1,j], sol[3*I-1,j-1], sol[3*I-1,j-2],
                                phiC[j], phiC[j-1], phiC[j-2], Dt, Dx[I-1], chi2 )
            
            
 #       if t_step_method == 1:
  #          
  #          current[0,j] = - ( sol[2*I,j] - sol[2*I,j-1] ) / (Dx[0] * Dt * chi2)
  #          current[1,j] = - ( phiC[j] - sol[3*I-1,j] - phiC[j-1] + sol[3*I-1,j-1] ) / (Dx[I-1] * Dt * chi2)
  #      elif t_step_method == 2:
  #          current[0,j] = - ( sol[2*I,j] - sol[2*I,j-1] ) / (Dx[0] * Dt * chi2)
  #          current[1,j] = - ( phiC[j] - sol[3*I-1,j] - phiC[j-1] + sol[3*I-1,j-1] ) / (Dx[I-1] * Dt * chi2)
            

    t2 = clock()
    
    simtime = t2-t1

    print("Simulation Runtime: ", simtime)
    
    return current, simtime

def calcAnodicCurrent( phiN, phi1, phi2, Dt, Dx, chi2 ):

    # BDF 2
    current_A = -(3 * (phiN) - 4 * (phi1) + (phi2)) / (2 * Dt * Dx * chi2)

    return current_A 

def calcCatodicCurrent( phiN, phi1, phi2, phiCN, phiC1, phiC2, Dt, Dx, chi2 ):

    # BDF 2
    current_C = -(3 * (phiN - phiCN) - 4 * (phi1 - phiC1) + (phi2 - phiC2)) / (2 * Dt * Dx * chi2)

    return current_C

def calcAxis( I ):
    
    # create x axis for simulation
    xi = np.zeros(I+1, dtype = np.float64)
    x_ = np.zeros(I+1, dtype = np.float64)

    for i in range(0,I+1):

        xi[i] = i / I
 
    # creating x axis with functional relation
    #x_ = np.sin( ( np.pi * xi  ) / 2 ) ** 2
    x_ = xi
    
    # cell centers
    centers = np.empty(I, dtype = np.float64 )
    for i in range(I):

        centers[i] = x_[i] + ( x_[i+1] - x_[i] ) / 2
    
    # getting cell volumes
    Dx = np.zeros(I, dtype = np.float64)
    Dx = x_[1:] - x_[:I]

    return Dx, centers
```

```python
_ELE_CHARGE = 1.6021766208e-19 # [Coulomb], constant value
_AVOGADRO_CONST = 6.022140857e23 # [1/mol]
_EPSILON_VAC = 8.854187817e-12 # original value in [Farad/m],
_BOLTZMANN_CONST = 1.38064852e-23 # [Joule/Kelvin], constant value

# Setup
model = 0
lengthscale = 1e9
M = 1e0
testname = r"test_impedance_1.txt"

# calculate constants
T = 300 # in  [K]
L = 10 # in [nm]

c0_in = 1 # mol / l
epsilon_m = 80 # solvent electrical permittivity
D0_in = 1e-12 # reference diffusion constant in [m^2/s]

# factor, diffusion constants of anions and cations
DC = 1
DA = 1

# factor solvent permittivity
epsilon = 1



# reaction boundary condition
reaction_bc = {
            "kA": 0.0,
            "kC": 0.0,
            "foxA": 0.0,
            "foxC": 0.0
               }


# calculate numerical constants
D0 = D0_in * (lengthscale ) ** 2
c0 = c0_in * _AVOGADRO_CONST * 1e3 * ( lengthscale ) ** (-3)
phi0 = (_BOLTZMANN_CONST * T) / _ELE_CHARGE

#chi1 = st.calcChi1( phi0, T )
#chi2 = st.calcChi2( L, c0, phi0, epsilon_m )

chi1 = 1.0
chi2 = c0 * lengthscale * (_ELE_CHARGE * L) ** 2/ (epsilon_m * _EPSILON_VAC * _BOLTZMANN_CONST * T)
T0 = M * L ** 2 / D0
f0 = c0 * D0 / (M * L)
```

```python
### input voltage

# spatial resolution
I = 60

# initial condition
sol_initial = np.zeros([3*I], dtype = np.float64)
sol_initial[0:2*I] = 1.0 / 2.0

N = 2 ** 17
Dt = 1e-3

# create time axis
t = np.zeros(N, dtype = np.float64)
for j in range(0,N):
    
    t[j] = j * Dt
    
    
# frequency in Hz and amplitude [mV]
freq = 7e2
U0 = 10

phiC = (U0 * 1e-3) * np.sin( 2 * np.pi * freq * T0 * t) / phi0

fig = plt.figure(dpi = 130)

ax = fig.add_subplot(1,1,1)

ax.plot(t, phiC, color = "blue")

plt.show()

print(N)
```

```python
# Run simulation
current, simtime = solver( I, N, epsilon, DC, DA, sol_initial, chi2, M, reaction_bc )
```
