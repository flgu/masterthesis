```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import data

filename = r"test_notebook_Dt.txt"

freq, Z, Dt, simtime = np.loadtxt(filename, delimiter= ',', dtype = np.complex128, unpack = True)
```

```python
# Nyquist plot
fig = plt.figure(dpi = 120)

ax = fig.add_subplot(1,1,1)

ax.plot(Z.real, Z.imag, marker = ".", ls = "None")

fig.tight_layout()

plt.show()

# Bode plot
fig = plt.figure(dpi = 120)

ax = fig.add_subplot(1,1,1)

ax.plot(Dt, np.absolute(Z), marker = ".", ls = "None")

fig.tight_layout()

ax.set_xscale("log")
ax.set_yscale("log")

plt.show()
```

```python
print(np.std(Z.imag) / np.mean(Z.imag) * 100)
print(np.std(Z.real) / np.mean(Z.real) * 100)
```
