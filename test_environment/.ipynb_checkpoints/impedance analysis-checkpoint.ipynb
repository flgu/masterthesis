```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import data

filename = r"model0_series0.txt"

freq, Z = np.loadtxt(filename, delimiter= ',', dtype = np.complex128, unpack = True)
```

```python
# Nyquist plot
fig = plt.figure(dpi = 120)

ax = fig.add_subplot(1,1,1)

ax.plot(Z.real, Z.imag, marker = ".", ls = "None")

fig.tight_layout()

plt.show()

# Bode plot
fig = plt.figure(dpi = 120)

ax = fig.add_subplot(1,1,1)

ax.plot(freq, np.absolute(Z), marker = ".", ls = "None")

fig.tight_layout()

ax.set_xscale("log")
ax.set_yscale("log")

plt.show()
```
